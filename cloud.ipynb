```python
import requests
from bs4 import BeautifulSoup
import bs4
import os
import re
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import jieba
from wordcloud import WordCloud, STOPWORDS
d='D:\\FFOutput'
os.chdir('D:\\FFOutput')

#################################################
ulist = []
url = 'http://www.chinanews.com/cj/2020/06-28/9223366.shtml'
r = requests.get(url, timeout=30)
r.raise_for_status()
r.encoding = "UTF-8"
html=r.text
soup = BeautifulSoup(html, "html.parser",from_encoding="utf-8")  #html
ct=soup.find_all('p')#<p>
ct=str(ct)
res=re.findall('<p>(.*?)</p>',ct)#<>
res=str(res)
red=re.findall('(.*?)</a>',res)
red=str(red)
dr = re.compile(r'<[^>]+>',re.S)#<re.compile
dd = dr.sub('',red)
dd=str(dd)
cc=''.join(dd.split('u3000'))#u3000

with open('data.txt','w') as f:    #
    f.write(cc)#


##################################################
    d = path.dirname('D:\\FFOutput')
# Read the whole text.
file = open(path.join(d, 'data.txt')).read()
##jieba
default_mode =jieba.cut(file)
text = " ".join(default_mode)
alice_mask = np.array(Image.open(path.join(d, "bad.jpg")))#
stopwords = set(STOPWORDS)
stopwords.add("said")
wc = WordCloud(  
    font_path='wqy-microhei.ttc',  #
    background_color="white",   #
    max_words=20000,     #
    mask=alice_mask,  
    stopwords=stopwords)  
wc.generate(text)
wc.to_file(path.join(d, "0234.jpg"))

##################################
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
