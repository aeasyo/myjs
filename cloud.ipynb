```python
import requests
from bs4 import BeautifulSoup
import bs4
import os
import re
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import jieba
from wordcloud import WordCloud, STOPWORDS
d='D:\\FFOutput'
os.chdir('D:\\FFOutput')

#####################爬虫部分############################
ulist = []
url = 'http://www.chinanews.com/cj/2020/06-28/9223366.shtml'
r = requests.get(url, timeout=30)
r.raise_for_status()
r.encoding = "UTF-8"
html=r.text
soup = BeautifulSoup(html, "html.parser",from_encoding="utf-8")  #解析html
ct=soup.find_all('p')#提取标签<p>
ct=str(ct)
res=re.findall('<p>(.*?)</p>',ct)#<用正则表达式去除标签>
res=str(res)
red=re.findall('(.*?)</a>',res)
red=str(red)
dr = re.compile(r'<[^>]+>',re.S)#<用re.compile函数去除标签
dd = dr.sub('',red)
dd=str(dd)
cc=''.join(dd.split('u3000'))#去除u3000

with open('data.txt','w') as f:    #设置文件对象
    f.write(cc)#写入文档


#####################词云生成部分#############################
    d = path.dirname('D:\\FFOutput')
# Read the whole text.
file = open(path.join(d, 'data.txt')).read()
##用jieba进行分词
default_mode =jieba.cut(file)
text = " ".join(default_mode)
alice_mask = np.array(Image.open(path.join(d, "bad.jpg")))#指定词云图片模版
stopwords = set(STOPWORDS)
stopwords.add("said")
wc = WordCloud(  
    font_path='wqy-microhei.ttc',  #设置字体，不指定就会出现乱码
    background_color="white",   #背景颜色
    max_words=20000,     #最大词汇数
    mask=alice_mask,  
    stopwords=stopwords)  
wc.generate(text)
wc.to_file(path.join(d, "0234.jpg"))

###############显示部分###################
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
