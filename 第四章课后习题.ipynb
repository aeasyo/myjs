```python
from random import randrange
values=[randrange(1,50) for i in range(20)]
print(values)
def insert_sort(data):

        for i in range(1, len(data)):
            value = data[i]  # 插入值
            
            j = i-1 # 记录插入位置  j+1 代表插入的位置
            while j >= 0:    # 不用 for 循环是因为 j 为 -1 的情况
                if  data[j] > value:
                    data[j+1] = data[j]
                else :
                    break
                j -= 1

            data[j+1] = value
        return data
```

```python
insert_sort(values)
```

```python
from random import randint

result = set()
while True:
    result.add(randint(1,10)) #修改范围
    if len(result) == 20:
        break
print(result)
```

```python
#result为集合，randint(1,10)只能产生从1到10的10个不重复数字，集合长度达不到20，因此死循环
```
